### Importing and cleaning the data

```python
data=pd.read_csv("bigmart_data.csv")
data=data.dropna(how='any')
data.head()
```

## Line chart
Creating a line chart to check the mean price per item

```python
#Calculating the mean of the price per item

Price_by_item= data.groupby("Item_Type").Item_MRP.mean()[:10]
Price_by_item
```

```python
#coverting the x and y axis values to a list
x=Price_by_item.index.tolist()
y=Price_by_item.values.tolist()

plt.figure(figsize=(14,8))

plt.title("Mean Prize for each item type")

plt.xlabel("Item_Type")
plt.ylabel("Mean Price")

plt.plot(x, y)
```

## Bar Chart
Creating a bar chart to look at the mean sales for each outlet size

```python
#Grouping the mean of the outlet sales by the outle size
Sales_by_outlet_size= data.groupby('Outlet_Size').Item_Outlet_Sales.mean()
Sales_by_outlet_size
```

```python
#Sorting the values making sure it is in ascending order
Sales_by_outlet_size.sort_values(inplace=True)
```

```python
Sales_by_outlet_size
```

```python
#coverting the x and y axis values to a list
X= Sales_by_outlet_size.index.tolist()
Y= Sales_by_outlet_size.values.tolist()
```

```python
#labeling the X and Y axis
plt.xlabel("Outlet Size")
plt.ylabel("Mean sales")

#Seting title for the plot
plt.title("Mean Sales for each outlet")

#Setting the labels for X
plt.xticks(labels=X, ticks=np.arange(len(X)))

#Ploting the bar chart
plt.bar(X, Y, color=["red", "purple", "magenta"])
```

## Histogram
To check the distribution of item price

```python
plt.title("Item MRP (price) distribution")

#xlabel
plt.xlabel("Item_MRP")

#ylabel
plt.ylabel("Frequency")

#Plotting the histogram
plt.hist(data["Item_MRP"], bins=20, color="lightblue")
```

## Box Plot
To check the distribution of sales

```python
data_ds=data[["Item_Outlet_Sales"]]

#Creating outlier point shape
red_diamond= dict(markerfacecolor= "r", marker= "D")

#Setting title
plt.title('Item Sales Distribution')

plt.boxplot(data_ds.values, labels=["Item Sales"], flierprops=red_diamond)
```

```python
data_ds=data[["Item_Weight", "Item_MRP"]]

#Creating outlier point shape
red_diamond= dict(markerfacecolor= "r", marker= "D")

fig, ax=plt.subplots()

plt.boxplot(data_ds.values, labels=["Item Weight", "Item MRP (price)"], flierprops=red_diamond)
```

## Violin Plots
For Density Distribution

```python
#Visualising the density distriution of item weights and item price

#Generating the subplot
fig, ax = plt.subplots()

#Adding label to the X axis
plt.xticks(ticks=[1,2], labels=["Item weight", "Item MRP"])

#plot the violinplot
plt.violinplot(data_ds.values)
```

## Scatter Plot

```python
#To calculate relative distribution of item weight and it's visibility

#Setting labels for those axes
plt.xlabel("Item Weight")
plt.ylabel("Item Visibility")

plt.scatter(data["Item_Weight"][:200],data["Item_Visibility"][:200])
```

## Bubble Plots
It helps to understand the interdependent relations among 3 variables

```python
plt.xlabel("Item_MRP")
plt.ylabel("Item_Outlet_Sales")

plt.title("Item Outlet Sales vs Item MRP (Price)")

plt.scatter(data["Item_MRP"][:100],data["Item_Outlet_Sales"][:100], s=data["Item_Visibility"][:100]*1000)
```
