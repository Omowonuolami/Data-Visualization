## Importing the libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
sns.set()
sns.set(style="darkgrid")
```

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

## Importing the data

```python
data=pd.read_csv("bigmart_data.csv")
data= data.dropna(how='any')

data["Visibility_Scaled"]= data["Item_Visibility"]*100
data.head()
```

## Creating Basic Plots

Line Chart

```python
sns.lineplot(x="Item_Weight", y="Item_MRP", data=data[:50]);
```

Bar Chart

```python
sns.barplot(x="Item_Type", y="Item_MRP", data=data[:5])
```

Histogram

```python
sns.distplot(data["Item_MRP"])
```

Box plot

```python
sns.boxplot(data["Item_Outlet_Sales"], orient="vertical")
```

Violin Plot

```python
sns.violinplot(data["Item_Outlet_Sales"], orient='vertical', color= "lightgreen")
```

```python
sns.relplot(x="Item_MRP", y="Item_Outlet_Sales", data= data[:200], kind= "scatter")
```

```python
sns.relplot(x="Item_MRP", y="Item_Outlet_Sales", hue="Item_Type", data= data[:200], kind= "scatter")
```

```python
#different line plot for differnt categories of the outlet size
sns.lineplot(x='Item_Weight', y='Item_MRP', hue='Outlet_Size', data=data[:150])
```

Bubble Plot

```python
sns.relplot(x="Item_MRP", y="Item_Outlet_Sales", data=data[:200], kind="scatter", size="Visibility_Scaled", hue="Visibility_Scaled")
```

### Categorical Scatterplot
Strip Plot

```python
sns.catplot(x="Outlet_Size", y="Item_Outlet_Sales",kind="strip", data=data[:250])
```

Swarm Plot

```python
sns.catplot(x="Outlet_Size", y="Item_Outlet_Sales",kind="swarm", data=data[:250])
```

Box Plot

```python
sns.catplot(x="Outlet_Size", y="Item_Outlet_Sales",kind="box", data=data)
```

Violin Plot

```python
sns.catplot(x="Outlet_Size", y="Item_Outlet_Sales",kind="violin", data=data)
```

Bar Plot

```python
sns.catplot(x="Outlet_Size", y="Item_Outlet_Sales",kind="bar", data=data)
```

### Density Plots

```python
#Plotting the distribution of item MRP
plt.figure(figsize=(7,7))
sns.kdeplot(data["Item_MRP"], shade=True);
```

```python
#Plotting the distribution of item visibility
plt.figure(figsize=(7,7))
sns.kdeplot(data["Item_Visibility"], shade=True);
```

Histogram and Density Plot

```python
plt.figure(figsize=(7,7))
sns.distplot(data["Item_Outlet_Sales"]);
```
